```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv(r"C:\Users\TEJA\Desktop\BeijingPM20100101_20151231.csv")
```

```python
#remove some unimportant columns
data.drop(['PM_Dongsi', 'PM_Dongsihuan', 'PM_Nongzhanguan','No'],axis=1,inplace=True)
```

```python
data["pm2.5"]=data["PM_US Post"]
data.drop(["PM_US Post"],axis=1,inplace=True)
data=data.dropna()
```

```python
data.groupby('hour')['pm2.5'].mean().plot.bar()
```

```python
data.groupby('month')['pm2.5'].mean().plot.bar()
```

```python
data.groupby('day')['pm2.5'].mean().plot.bar()
```

from the above graphs I found some important features which i grouped and made a separate column which i thought would be useful.

but it decreasing the accuracy.so i didn't added it in the model.

data["hour"]=data["hour"].map(lambda s:1 if s==0 or s==1 or s==2 or s==3 or s==20 or s==21 or s==22 or s==23 else 0

data["day"]=data["day"].map(lambda s:1 if s==9 or s==15 or s==16 or s==18 or s==19 or s==20 or s==29  else 0)

data["month"]=data["month"].map(lambda s:1 if s==2 or s==6 or s==7 or s==10 or s==11 or s==12   else 0)

the date and time in the above conditions will have high pm2.5

data["more_pm2.5_time"]=data["hour"]+data["day"]+data["month"]

data["more_pm2.5_time"]=data["more_pm2.5_time"].map(lambda s:1 if s==3 else 0)



```python
data["season"].value_counts()

data.groupby('season')['pm2.5'].mean().plot.bar()

#all the seasons carry important priority some one hot encode the seasons
```

```python
season_dummies=pd.get_dummies(data["season"])
data=pd.concat([data,season_dummies],axis=1)
data.drop(["season"],1,inplace=True)
```

```python
cbwd_dummies=pd.get_dummies(data["cbwd"])
data=pd.concat([data,cbwd_dummies],axis=1)
data.drop(["cbwd"],1,inplace=True)
```

```python
year_dummies=pd.get_dummies(data["year"])
data=pd.concat([data,year_dummies],axis=1)
data.drop(["year"],1,inplace=True)
```

```python
#normalize the numerical values

data["PRES"]=(data["PRES"]-np.min(data["PRES"]))/(np.max(data["PRES"])-np.min(data["PRES"]))
data["DEWP"]=(data["DEWP"]-np.min(data["DEWP"]))/(np.max(data["DEWP"])-np.min(data["DEWP"]))
data["TEMP"]=(data["TEMP"]-np.min(data["TEMP"]))/(np.max(data["TEMP"])-np.min(data["TEMP"]))
data["HUMI"]=(data["HUMI"]-np.min(data["HUMI"]))/(np.max(data["HUMI"])-np.min(data["HUMI"]))
data["Iws"]=(data["Iws"]-np.min(data["Iws"]))/(np.max(data["Iws"])-np.min(data["Iws"]))
```

```python
data["pm2.5"]=data["pm2.5"].map(lambda s:1 if s>115 else 0)
Y=data["pm2.5"]
data.drop(["pm2.5"],1,inplace=True)

data=data.dropna()
```

```python
data.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, Y, test_size=0.25,random_state=50)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)

print('Accuracy : {:.2f}'.format(logreg.score(X_test, y_test)))
```

RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

model=RandomForestClassifier(n_estimators=10,random_state=50)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)


print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
data=pd.read_csv(r"C:\Users\TEJA\Desktop\PRSA_data_2010.1.1-2014.12.31.csv")
```

```python
#convert the date and time into the standard form

column=str("-")
data["modified_time"]= data['year'].astype(str)+column+data['month'].astype(str)+column+data['day'].astype(str) 
data["modified_time"]=pd.to_datetime(data["modified_time"], format = '%Y-%m-%d')
```

```python
ar_data=data.iloc[:,[13,5]]
ar_data.set_index('modified_time', inplace=True)
ar_data=ar_data.dropna()
```

```python
ar_data_days = ar_data.groupby('modified_time')
modified_pm2_day=ar_data_days.agg(np.mean)

plt.title("pm2.5 variation with respective dates")
plt.plot(modified_pm2_day)
```

```python
X_day = modified_pm2_day.values
train=X_day[:1780]
test=X_day[1780:]
```

```python
from statsmodels.tsa.ar_model import AR
from sklearn.metrics import mean_squared_error
model_ar = AR(train)
model_ar_fit = model_ar.fit()

predictions = model_ar_fit.predict(start=1780,end=1788)


plt.title("pm2.5 actual vs predicted using daily data")
plt.plot(test)
plt.plot(predictions,color='red')
```
